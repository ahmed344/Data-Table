```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu, median_abs_deviation

# statistical annotation
def p_annotation(p, x1, x2, y, h=0.005, color='k',linewidth=3, font=25):
    plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=linewidth, c=color)
    plt.text((x1+x2)*.5, y+h, f'P = {p:.2f}', ha='center', va='bottom', color=color, fontsize=font)
```

### Read and clean the data

```python
# Read the csv
table_nta = pd.read_csv('Data/Cells_on_BSA_Experiments.csv').iloc[:9]
table_bsa = pd.read_csv('Data/Cells_on_BSA_Experiments.csv').iloc[[10,11,25,26,27,28]]
table_bio = pd.read_csv('Data/Cells_on_BSA_Experiments.csv').iloc[29:]
table_slb = pd.read_csv('Data/Cells_Experiments.csv')

# Drop columns
table_nta = table_nta.drop(['Date', 'Observation', 'Undefined_Ad', 'Small_Bl', 'Big_Bl', 'Enormous_Bl', 'Notes'], axis=1)
table_bsa = table_bsa.drop(['Date', 'Observation', 'Undefined_Ad', 'Small_Bl', 'Big_Bl', 'Enormous_Bl', 'Notes'], axis=1)
table_bio = table_bio.drop(['Date', 'Observation', 'Undefined_Ad', 'Small_Bl', 'Big_Bl', 'Enormous_Bl', 'Notes'], axis=1)
table_slb = table_slb.drop(['Date', 'Observation'], axis=1)

# Rename columns
table_nta = table_nta.rename(columns={'Discription':'Sample', 'Incubation':'Functionalization', 'No_Ad':'No', 'Small_Ad':'Small', 'Big_Ad':'Big'})
table_bsa = table_bsa.rename(columns={'Discription':'Sample', 'Incubation':'Concentration'    , 'No_Ad':'No', 'Small_Ad':'Small', 'Big_Ad':'Big'})
table_bio = table_bio.rename(columns={'Discription':'Sample', 'Incubation':'Functionalization', 'No_Ad':'No', 'Small_Ad':'Small', 'Big_Ad':'Big'})
table_slb = table_slb.rename(columns={'Discription':'Sample', 'Incubation':'Functionalization', 'No_Ad':'No', 'Small_Ad':'Small', 'Big_Ad':'Big'})

# Augment the small by 1 for BSA-NTA
table_nta['Small'] +=1

# Adhesion ratio
table_nta['Rbig'  ] = table_nta['Big'] / (table_nta['No'] + table_nta['Small'])
table_bsa['Rsmall'] = (table_bsa['Big'] + table_bsa['Small']) / table_bsa['No']
table_bio['Rsmall'] = (table_bio['Big'] + table_bio['Small']) / table_bio['No']
table_slb['Rsmall'] = (table_slb['Big'] + table_slb['Small']) / table_slb['No']

# print the tables in latex format
print(table_nta.round(2).to_latex(index=False))
print(table_bsa.round(2).to_latex(index=False))
print(table_bio.round(2).to_latex(index=False))
print(table_slb.round(2).to_latex(index=False))
```

### The Mann-Whitney U test P-value for SLB

```python
# Filter the data
nta_p   = table_nta[table_nta.Functionalization.str.contains("with "   )].Rbig.values
nta_n   = table_nta[table_nta.Functionalization.str.contains("without ")].Rbig.values

bsa_1   = table_bsa[table_bsa.Concentration.str.contains(" 1mg")].Rsmall.values
bsa_5   = table_bsa[table_bsa.Concentration.str.contains(" 5mg")].Rsmall.values

bio_p   = table_bio[table_bio.Functionalization.str.contains("with "   )].Rsmall.values
bio_n   = table_bio[table_bio.Functionalization.str.contains("without ")].Rsmall.values

sopc_5  = table_slb[table_slb.Functionalization.str.contains("with "   )][table_slb.Sample.str.contains("SOPC"      )].Rsmall.values
sopc_0  = table_slb[table_slb.Functionalization.str.contains("without ")][table_slb.Sample.str.contains("SOPC"      )].Rsmall.values

dppc_5  = table_slb[table_slb.Functionalization.str.contains("with "   )][table_slb.Sample.str.contains("DPPC 5NTA" )].Rsmall.values
dppc_10 = table_slb[table_slb.Functionalization.str.contains("with "   )][table_slb.Sample.str.contains("DPPC 10NTA")].Rsmall.values
dppc_p  = table_slb[table_slb.Functionalization.str.contains("with "   )][table_slb.Sample.str.contains("DPPC "     )].Rsmall.values
dppc_0  = table_slb[table_slb.Functionalization.str.contains("without ")][table_slb.Sample.str.contains("DPPC "     )].Rsmall.values

# Calculate the Mann-Whitney U test P-value
_, p_nta  = mannwhitneyu(nta_p, nta_n)

_, p_bsa  = mannwhitneyu(bsa_1, bsa_5)

_, p_bio  = mannwhitneyu(bio_p, bio_n)

_, p_sopc = mannwhitneyu(sopc_5, sopc_0)
_, p_dppc = mannwhitneyu(dppc_p, dppc_0)
_, p_with = mannwhitneyu(sopc_5, dppc_p)
_, p_wout = mannwhitneyu(sopc_0, dppc_0)

print(f'The Mann-Whitney U test P-value')
print(f'NTA (+) vs NTA (-) = {p_nta:.5f}')
print(f'bsa (1) vs BSA (5) = {p_bsa:.5f}')
print(f'BIO (+) vs BIO (-) = {p_bio:.5f}')
print(f'SOPC (+) vs SOPC (-) = {p_sopc:.5f}')
print(f'DPPC (+) vs DPPC (-) = {p_dppc:.5f}')
print(f'SOPC (+) vs DPPC (+) = {p_with:.5f}')
print(f'SOPC (-) vs DPPC (-) = {p_wout:.5f}')
```

### Box plots

```python
plt.figure(figsize=(12,10))     # 16.54,20

# Define the boxes
bp = plt.boxplot([bsa_1, bsa_5],
                labels = [f'BSA 1 mg/mL \n  $<R_s>=$ {bsa_1.mean():.2f}\n  $\sigma(R_s)$   = {bsa_1.std():.2f} \n $med(R_s)=$ {np.median(bsa_1):.2f}\n $MAD(R_s)=$ {median_abs_deviation(bsa_1):.2f}',
                          f'BSA 5 mg/mL \n  $<R_s>=$ {bsa_5.mean():.2f}\n  $\sigma(R_s)$   = {bsa_5.std():.2f} \n $med(R_s)=$ {np.median(bsa_5):.2f}\n $MAD(R_s)=$ {median_abs_deviation(bsa_5):.2f}'],
                widths = 0.5,
                patch_artist=True);

# Customize the boxes
bp['boxes'][0].set(color='green', hatch='/', linewidth=4, alpha=0.4)
bp['boxes'][1].set(color='brown', hatch='/', linewidth=4, alpha=0.4)


# Customize the Median line
bp['medians'][0].set(color='green', linewidth=4, alpha=0.6)
bp['medians'][1].set(color='brown', linewidth=4, alpha=0.6)


# Plot the scattered points
plt.scatter(x = [1] * len(bsa_1) , y = bsa_1 , color='green', s=100, alpha=0.8, label='BSA 1 mg/mL' )
plt.scatter(x = [2] * len(bsa_5) , y = bsa_5 , color='brown', s=100, alpha=0.8, label='BSA 5 mg/mL')


# # statistical annotation
p_annotation(p_bsa, 1, 2, bsa_1.max() + 0.1, h=0.03)

# Customize the table
plt.ylabel('$R_s$', fontsize= 'xx-large')
plt.xticks(fontsize= 'xx-large')
plt.yticks(fontsize= 'xx-large')
plt.ylim(( -0.1,1.5))
plt.yscale('symlog')

# plt.title('Reduced area with different cadherin concentration on fluid and fixed SLBs', fontsize=23)
plt.grid(alpha=0.5)
#plt.legend(fontsize='xx-large', loc='upper left')

# Save the figure
plt.savefig("adhesion_ratio_bsa", bbox_inches = 'tight')
plt.show()
```
